```python
x = []
counter = 0
for row in train.iterrows():
    counter += 1
    if counter % 1000 == 0:
        print(counter)
    x.append(list(train_features.loc[[row[1]['parcelid']]].values[0]))
    
```

```python
data = pd.DataFrame(x, columns = train_features.columns)
```

```python
categry_columns = [x.replace('\'', '') for x in list(train_labels[train_labels['Type'] == 'c']['Feature'].values) if x != '\'parcelid\'']
```

```python
processed_data = pd.get_dummies(data, columns = categry_columns)
```

```python
import numpy as np
from sklearn import linear_model
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn import cross_validation
```

```python
counter = 0
for row in train_labels.iterrows():
    try:
        data_type = row[1]['Type']
        if 'parcelid' in row[1]['Feature']:
            continue
        print( row[1]['Feature'])
        if data_type == 'c':
            feature = data[row[1]['Feature'].replace("'", '')].fillna(0).values.reshape(-1, 1)
            enc = OneHotEncoder()
            enc.fit(feature)
            d = enc.transform(feature).toarray()
            reg = linear_model.LinearRegression()
            reg.fit(d, train['logerror'].values)
        else:
            feature = data[row[1]['Feature'].replace("'", '')].fillna(0).values.reshape(-1, 1)
            reg.fit(feature, train['logerror'].values)
            print(feature[:5], train['logerror'].values[:5])
        print(reg.residues_)
        print(reg.coef_)
    except ValueError:
        break
        
    counter += 1
```

```python
model = RandomForestRegressor(n_jobs=10, n_estimators=30, max_features='auto')
```

```python
cross_validation.cross_val_score(model, processed_data, train['logerror'].values, cv=5)
```
