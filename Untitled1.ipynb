## Compare continuous variables pairwise:

```python
numerical_fields = [x.replace('\'', '') for x in train_labels[train_labels['Type'] == 'n']['Feature'].values]
```

```python
numerical_field_number = len(numerical_fields)
```

```python
fields_of_interst = ['bathroomcnt', 'bedroomcnt', 'buildingqualitytypeid', 'finishedsquarefeet12', 
                    'lotsizesquarefeet', 'structuretaxvaluedollarcnt', 'taxvaluedollarcnt', 'landtaxvaluedollarcnt', 'taxamount']
```

```python
numerical_fields = fields_of_interst
numerical_field_number = len(numerical_fields)
```

```python
for i in range(numerical_field_number):
    for j in range(numerical_field_number):
        if i == j:
            continue
        df_not_null = train_large_error[(train_large_error[numerical_fields[i]].notnull() & train_large_error[numerical_fields[j]].notnull())]
        df_not_zero = df_not_null[(df_not_null[[numerical_fields[i], numerical_fields[j]]] != 0).all(1)].copy()
        if len(df_not_zero) == 0:
            continue
        for index in [i, j]:
            df_not_zero[numerical_fields[index]] = preprocessing.normalize(df_not_zero[numerical_fields[index]], norm='l2').reshape(-1, 1) 
        thres = np.percentile(df_not_zero[numerical_fields[i]] / df_not_zero[numerical_fields[j]], 85)
        df_not_zero = df_not_zero[df_not_zero[numerical_fields[i]] / df_not_zero[numerical_fields[j]] < thres]
        if len(df_not_zero) == 0:
            continue
        # do scatter plot here
        x = sm.add_constant(df_not_zero['logerror']) # constant intercept term
        y = df_not_zero[numerical_fields[i]] / df_not_zero[numerical_fields[j]]
        model = sm.OLS(y, x)
        fitted = model.fit()
        x_pred = np.linspace(df_not_zero['logerror'].min(), df_not_zero['logerror'].max(), 50)
        x_pred2 = sm.add_constant(x_pred)
        y_pred = fitted.predict(x_pred2)
        
        y_hat = fitted.predict(x) 
        y_err = y - y_hat
        mean_x = x['logerror'].T.mean()
        n = len(x)
        dof = n - fitted.df_model - 1
        t = stats.t.ppf(1-0.025, df=dof)
        s_err = np.sum(np.power(y_err, 2))
        conf = t * np.sqrt((s_err/(n-2))*(1.0/n + (np.power((x_pred-mean_x),2) / ((np.sum(np.power(x_pred,2))) - n*(np.power(mean_x,2))))))
        upper = y_pred + abs(conf)
        lower = y_pred - abs(conf)
        
        sdev, lower, upper = wls_prediction_std(fitted, exog=x_pred2, alpha=0.05)
            
        
        fig, ax = plt.subplots(figsize=(5, 5))
        ax.scatter(df_not_zero['logerror'], y, alpha=0.5, color='orchid')
        ax.plot(x_pred, y_pred, '-', color='darkorchid', linewidth=2)
        ax.fill_between(x_pred, lower, upper, color='#888888', alpha=0.4)
        ax.fill_between(x_pred, lower, upper, color='#888888', alpha=0.1)
        ax.set_title(' '.join([numerical_fields[i], numerical_fields[j]]))
        
        
```
